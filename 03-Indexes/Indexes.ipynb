```python
people = {
    "first": ["Corey", 'Jane', 'John'],
    "last": ["Schafer", 'Doe', 'Doe'],
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
df["email"]    # Acessando a coluna email
```

```python
df.index
```

```python
# Define o índice para a coluna email e modifica o DataFrame original
df.set_index("email", inplace=True)
# o inplace=True é necessário para modificar o DataFrame original, caso contrário,
# ele retornaria um novo DataFrame com o índice definido, mas não modificaria o DataFrame original.
```

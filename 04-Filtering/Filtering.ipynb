```python
people = {
    "first": ["Corey", 'Jane', 'John'],
    "last": ["Schafer", 'Doe', 'Doe'],
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
filtro = (df['last'] == 'Doe' ) # Retorna una Serie de valores booleanos indicando se o valor na coluna 'last' é igual a 'Doe' para cada linha do DataFrame.
```

```python
df[filtro] # Retorna um DataFrame contendo apenas as linhas onde a condição do filtro é verdadeira, ou seja, onde a coluna 'last' é igual a 'Doe'.
```

```python
df[df['last'] == 'Doe'] # Retorna um DataFrame contendo apenas as linhas onde a coluna 'last' é igual a 'Doe', sem a necessidade de criar uma variável intermediária para o filtro.
```

```python
df.loc[filtro] # Usando a indexação .loc para obter o mesmo resultado do filtro, retornando um DataFrame contendo apenas as linhas onde a coluna 'last' é igual a 'Doe'.
```

```python
df.loc[filtro, 'email'] # Retorna apenas a coluna 'email' das linhas onde a coluna 'last' é igual a 'Doe',
# usando a indexação .loc para filtrar as linhas e selecionar a coluna específica.
```

```python
filtro2 = (df['last'] == 'Doe') & (df['first'] == 'John') # Combina dois filtros usando o operador lógico AND (&) para criar
# um filtro que retorna True apenas para as linhas onde a coluna 'last' é igual a 'Doe' e a coluna 'first' é igual a 'John'.

# & = AND
# | = OR
# ~ = NOT
```

```python
df[filtro2]
```

```python
df.loc[filtro2,'email'] # Retorna apenas a coluna 'email' das linhas onde ambas as condições do filtro são verdadeiras.
```

```python
df.loc[~filtro2] # Retorna a coluna 'email' das linhas onde a condição do filtro2 é falsa,
# ou seja, onde a coluna 'last' não é igual a 'Doe' ou a coluna 'first' não é igual a 'John'.
```

```python
df.loc[~filtro2, 'email'] # Retorna apenas a coluna 'email' das linhas onde a condição do filtro2 é falsa, usando a indexação
# .loc para filtrar as linhas e selecionar a coluna específica.
```

```python
people = {
    "first": ["Corey", 'Jane', 'John'],
    "last": ["Schafer", 'Doe', 'Doe'],
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
df.columns
```

```python
df.columns = ['first_name','last_name', 'email']
```

```python
df
```

```python
df.columns = [x.upper() for x in df.columns]
df
```

```python
df.columns = df.columns.str.replace('_',' ')
df
```

```python
df.columns = df.columns.str.replace(' ','_')
df.columns = [x.lower() for x in df.columns]
```

```python
df
```

```python
df.rename(columns={'first_name':'first', 'last_name': 'last'}, inplace = True)
df
```

```python
df.loc[2] = ['John','Smith','JhonSmith@email.com']
df
```

```python
df.loc[2, ['last','email']]
```

```python
df.loc[2, ['last','email']] = ['Matheus', 'Matheus@email.com']
df
```

```python
df.loc[2,'last'] = 'Carlos'
df
```

```python

# Usando o método at para atualizar um valor específico
# O método at é mais rápido do que loc para acessar um único valor, pois é otimizado para esse tipo de operação.
df.at[2,'last'] = 'Matheus'
df
```

```python
filtro = (df['email'] == 'JaneDoe@email.com')
df[filtro]['last'] = 'Smith'
df

# Essa operação não funciona porque o filtro retorna uma cópia do DataFrame, e não uma visão.
# Portanto, quando tentamos atualizar a coluna 'last' usando o filtro, estamos atualizando a cópia, e não o DataFrame original.
# Para atualizar o DataFrame original, precisamos usar loc ou at para acessar as linhas correspondentes ao filtro e atualizar a coluna 'last' diretamente no DataFrame original.
```

```python
filtro = (df['email'] == 'JaneDoe@email.com')
df.loc[filtro, 'last'] = 'Smith'
df
```

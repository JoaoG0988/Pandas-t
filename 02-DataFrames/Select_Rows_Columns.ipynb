```python
people = {
    "first": ["Corey"],
    "last": ["Schafer"],
    "email": ["CoreyMSchafer@gmail.com"]
}
```

```python
# As chaves são as colunas e os valores são as linhas, ou seja, os dados de cada pessoa.

people = {
    "first": ["Corey", 'Jane', 'John'],
    "last": ["Schafer", 'Doe', 'Doe'],
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}
```

```python
people["email"] # Acessa a coluna "email" do dicionário 'people' e retorna os valores associados a essa chave, que são os endereços de email das pessoas.
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people) # Cria um DataFrame a partir do dicionário 'people'.
# O DataFrame é uma estrutura de dados bidimensional, semelhante a uma tabela,
# que pode conter diferentes tipos de dados (como números, strings, etc.)
# e é amplamente utilizado para análise de dados em Python.
```

```python
df
```

```python
df["email"] # Acessa a coluna "email" do DataFrame 'df' e retorna os valores associados a essa coluna, que são os endereços de email das pessoas.
# É o que se chama de Series, que é uma estrutura de dados unidimensional do Pandas, semelhante a uma coluna em uma tabela, que pode conter diferentes tipos de dados e é indexada.
```

```python
df.email # Acessa a coluna "email" do DataFrame 'df' usando a notação de ponto, que é uma forma alternativa de acessar colunas em um DataFrame do Pandas.
# No entanto, essa notação só funciona se o nome da coluna for um identificador válido (sem espaços, caracteres especiais, etc.)
# e não for o mesmo que um método ou atributo do DataFrame.
```

```python
type(df["email"]) # Verifica o tipo do objeto retornado ao acessar a coluna "email" do DataFrame 'df', que é uma Series do Pandas.
```

```python
df[["last", "email"]] # Acessa as colunas "last" e "email" do DataFrame 'df' e retorna um novo DataFrame contendo apenas essas colunas.
```

```python
df.columns # Retorna um objeto Index contendo os rótulos das colunas do DataFrame 'df', que são "first", "last" e "email".
```

Funções iloc e loc

```python
df.iloc[0] # Acessa a primeira linha do DataFrame 'df' usando o método iloc, que é baseado em índices inteiros.
# Retorna uma Series contendo os valores da primeira linha.
```

```python
df.iloc[[0, 1]] # Acessa as primeiras duas linhas do DataFrame 'df' usando o método iloc e retorna um novo DataFrame contendo essas linhas.
```

```python
df.iloc[[0, 1], 2] # Acessa as primeiras duas linhas e a terceira coluna (índice 2) do DataFrame 'df'
# usando o método iloc e retorna os valores correspondentes, que são os endereços de email das duas primeiras pessoas.
```

```python
df.iloc[[0, 1], [2, 0]] # Acessa as primeiras duas linhas e as colunas de índice 2 e 0 do DataFrame 'df'
# usando o método iloc e retorna os valores correspondentes,
# que são os endereços de email e os primeiros nomes das duas primeiras pessoas.

# É possível mudar a ordem das colunas e linhas usando o método iloc, basta especificar os índices na ordem desejada.
```

```python
df.loc[[0,1], ["email", "first"]] # Acessa as primeiras duas linhas e as colunas "email" e "first" do DataFrame 'df' usando o método loc, que é baseado em rótulos.
# Retorna os valores correspondentes, que são os endereços de email e os primeiros
```
